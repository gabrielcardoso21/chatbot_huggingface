```python
# input sequence
text = "Quem era Jim Henson? Jim Henson era um"
inputs = tokenizer(text, return_tensors="pt")

# model output using Top-k sampling text generation method
sample_outputs = model.generate(inputs.input_ids,
                                pad_token_id=50256,
                                do_sample=True, 
                                max_length=90, # put the token number you want
                                top_k=40,
                                num_return_sequences=1)

# generated sequence
for i, sample_output in enumerate(sample_outputs):
    print(">> Generated text {}\n\n{}".format(i+1, tokenizer.decode(sample_output.tolist())))
```

```python
from transformers import AutoTokenizer, AutoModelWithLMHead, AutoModelForSeq2SeqLM, pipeline
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
import torch

# chatbot (en)
model_name = 'facebook/blenderbot-400M-distill'
tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
chatbot_model = BlenderbotForConditionalGeneration.from_pretrained(model_name)

#  Translator (en-pt)
en_pt_tokenizer = AutoTokenizer.from_pretrained("unicamp-dl/translation-en-pt-t5")
en_pt_model = AutoModelForSeq2SeqLM.from_pretrained("unicamp-dl/translation-en-pt-t5")
enpt_pipeline = pipeline('text2text-generation', model=en_pt_model, tokenizer=en_pt_tokenizer)

# Translator (pt-en)
pt_en_tokenizer = AutoTokenizer.from_pretrained("unicamp-dl/translation-pt-en-t5")
pt_en_model = AutoModelForSeq2SeqLM.from_pretrained("unicamp-dl/translation-pt-en-t5")
pten_pipeline = pipeline('text2text-generation', model=pt_en_model, tokenizer=pt_en_tokenizer)

# Question Answering
context = r"""
O Regime Próprio de Previdência Social é o sistema previdenciário para os servidores públicos efetivos da União, dos estados, do Distrito Federal e dos municípios.

Uma curiosidade importante para você: esse sistema é estabelecido separadamente para cada ente público. Isso significa que cada órgão tem seu Regime Próprio de Previdência.

O objetivo central desse Regime é organizar a previdência dos servidores públicos, titulares de cargo efetivo, excluindo os servidores comissionados, e incluindo as autarquias e fundações, para os ativos, aposentados e também de seus dependentes.
"""

model_name = 'pierreguillou/bert-base-cased-squad-v1.1-portuguese'
nlp = pipeline("question-answering", model=model_name)

# question = "Como faço pra aposentar?"

# result = nlp(question=question, context=context)

# print(f"score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}\nResposta: {result['answer']}")

def compute_answer(pt_input):
  result = nlp(question=pt_input, context=context)
  if result['score'] >= 0.2:
    return result['answer']
  in_txt = "translate Portuguese to English: %s" % pt_input
  en_input = pten_pipeline(in_txt)[0]["generated_text"]
  inputs = tokenizer(en_input, return_tensors="pt")
  result = chatbot_model.generate(**inputs)
  en_output = tokenizer.decode(result[0])[3:]
  en_output = en_output[:-4]
  pt_output = enpt_pipeline("translate English to Portuguese: %s" % en_output)[0]["generated_text"]
  return pt_output
```

```python
from difflib import SequenceMatcher
from thefuzz import fuzz
import pandas as pd
import numpy as np

ratcliff_overshelp = []
fuzzy = []

answers = {}

for min_score in min_score_values:
  answers[min_score] = {}
  
  y_resp = []
  for x in base_tests:
    answer = compute_answer(x, min_score)
    y_resp.append(answer)
    answers[min_score][x] = answer
  
  answers[min_score]['y_resp'] = y_resp

  erro_ratcliff = []
  erro_fuzzy = []
  for x in base_tests:
    erro_ratcliff.append(1 - SequenceMatcher(None,base_tests[x], answers[min_score][x]).ratio())
    erro_fuzzy.append(1 - fuzz.ratio(base_tests[x], answers[min_score][x])/100.0)

  answers[min_score]['erro_ratcliff'] = erro_ratcliff
  answers[min_score]['erro_fuzzy'] = erro_fuzzy

  ratcliff_overshelp += [max(erro_ratcliff), sum(erro_ratcliff)/len(erro_ratcliff), sum(np.square(erro_ratcliff)) /len(erro_ratcliff)]
  fuzzy += [max(erro_fuzzy), sum(erro_fuzzy)/len(erro_fuzzy), sum(np.square(erro_fuzzy)) /len(erro_fuzzy)]

header = pd.MultiIndex.from_product([min_score_values, ['Erro Máximo', 'Erro Médio', 'Erro Quadrático']], names=['Critério de decisão', 'Técnica \ Erro'])

pd.DataFrame([ratcliff_overshelp, fuzzy], index=['Ratcliff-Obershelp', 'Fuzzy'], columns=header)
```

```python
for min_score in min_score_values:
  res = {
      'Técnica': ['Ratcliff-Obershelp', 'Fuzzy'], 
      'Erro Máximo': [max(answers[min_score]['erro_ratcliff']), max(answers[min_score]['erro_fuzzy'])],
      'Erro Médio': [sum(answers[min_score]['erro_ratcliff'])/len(answers[min_score]['erro_ratcliff']), 
                     sum(answers[min_score]['erro_fuzzy'])/len(answers[min_score]['erro_fuzzy'])],
      'Erro Quadrático': [sum(np.square(answers[min_score]['erro_ratcliff'])) /len(answers[min_score]['erro_ratcliff']),
                          sum(np.square(answers[min_score]['erro_fuzzy'])) /len(answers[min_score]['erro_fuzzy'])]
      }
  print(pd.DataFrame(res).to_latex())
```
